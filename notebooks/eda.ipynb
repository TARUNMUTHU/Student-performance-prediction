```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset with semicolon delimiter
df = pd.read_csv('../data/student-mat.csv', sep=';')

# Convert all categorical columns to numeric (for correlation)
from sklearn.preprocessing import LabelEncoder
label_cols = df.select_dtypes(include='object').columns

for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Now check correlation matrix
corr = df.corr(numeric_only=True)

# Confirm its not empty
print("Correlation shape:", corr.shape)
print(corr.head())

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Student Performance Features")
plt.show()
```
